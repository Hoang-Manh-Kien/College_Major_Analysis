```python
import numpy as np
import pandas as pd
import matplotlib.ticker as ticker
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("D:\Kien Hoang\Kien Hoang\Data Files Folder\College_Major_Analysis\grad-students.csv")
data.head()
```

```python
data["Major_category"].unique().tolist()
```

```python
mean_grad_median = data.groupby(['Major_category', 'Major'])['Grad_median'].mean()

print(mean_grad_median)
```

```python
mean_grad_median_major_category = data.groupby(['Major_category'])['Grad_median'].mean()

print(mean_grad_median_major_category)
```

```python
sorted_data = sorted(mean_grad_median_major_category)
sorted_categories = [cat for _, cat in sorted(zip(mean_grad_median_major_category, mean_grad_median_major_category.index))]

plt.figure(figsize=(8, 6)) 

plt.barh(range(len(sorted_data)), sorted_data)

plt.xlabel("Median Salary of Grad ($)")
plt.yticks(range(len(sorted_data)), sorted_categories, size=8)

for i, count in enumerate(sorted_data):
    plt.text(count, i, f"{int(count//1000)}k", va='center')

plt.show()
```

```python
data.head()
```

```python
mean_nongrad_median = data.groupby(['Major_category', 'Major'])['Nongrad_median'].mean()
print(mean_nongrad_median)

mean_nongrad_median_major_category = data.groupby(['Major_category'])['Nongrad_median'].mean()
print(mean_nongrad_median_major_category)
```

```python
sorted_data_2 = sorted(mean_nongrad_median_major_category)
sorted_categories_2 = [cat for _, cat in sorted(zip(mean_nongrad_median_major_category, mean_nongrad_median_major_category.index))]

plt.figure(figsize=(8, 6)) 
plt.barh(range(len(sorted_data_2)), sorted_data_2)

plt.xlabel("Median Salary of Nongrad ($)")
plt.yticks(range(len(sorted_data_2)), sorted_categories_2, size=8)

for i, count in enumerate(sorted_data_2):
    plt.text(count, i, f"{int(count//1000)}k", va='center')

plt.show()
```

```python
difference = [grad - non_grad for grad, non_grad in zip(sorted_data, sorted_data_2)]

plt.figure(figsize=(9,7))

plt.barh(data["Major_category"].unique().tolist(), difference)
plt.xlabel("Difference in Median Salary ($)")
plt.title("Difference in Median Salary between Major Grads and Non-Grads")

for i, count in enumerate(difference):
    plt.text(count, i, f"{int(count//1000)}k", va='center')

mean_value = np.mean(difference)
plt.axvline(mean_value, color='red', linestyle='--', label='Mean Difference')

plt.show()

print("Mean:", mean_value)
```

```python
graduate_share = data['Grad_share']
graduate_premium = data['Grad_premium']

plt.figure(figsize=(7,5))

plt.scatter(graduate_share, graduate_premium)

plt.xlabel('Graduate Share (%)')
plt.ylabel('Graduate Premium')
plt.title('Relationship between Grad Share and Grad Premium')

plt.show()
```

```python
mean_grad_25 = data.groupby(['Major_category'])['Grad_P25'].mean()
print(mean_grad_25)

mean_grad_75 = data.groupby(['Major_category'])['Grad_P75'].mean()
print(mean_grad_75)
```

```python
categories = data["Major_category"].unique().tolist()
values1 = mean_grad_25
values2 = mean_grad_75

bar_width = 0.35
plt.figure(figsize=(8, 5)) 

x = np.arange(len(categories))

bar1 = plt.bar(x, values1, width=bar_width, label='Grad P25')
bar2 = plt.bar(x + bar_width, values2, width=bar_width, label='Grad P75')

plt.xlabel('Categories')
plt.xticks(rotation='vertical', size=8)
plt.ylabel('Income in 25th and 75th percentage ($)')

for bar in bar1 + bar2:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, f"{int(height/1000)}K", ha='center', va='bottom', fontsize=7)

plt.xticks(x + bar_width / 2, categories)

plt.legend(fontsize=7)
plt.show()
```

```python
mean_nongrad_25 = data.groupby(['Major_category'])['Nongrad_P25'].mean()
mean_nongrad_75 = data.groupby(['Major_category'])['Nongrad_P75'].mean()

categories = data["Major_category"].unique().tolist()
nongradvalues1 = mean_nongrad_25
nongradvalues2 = mean_nongrad_75

bar_width = 0.35
plt.figure(figsize=(8, 5)) 

x = np.arange(len(categories))

bar1 = plt.bar(x, nongradvalues1, width=bar_width, label='Nongrad P25')
bar2 = plt.bar(x + bar_width, nongradvalues2, width=bar_width, label='Nongrad P75')

plt.xlabel('Categories')
plt.xticks(rotation='vertical', size=8)
plt.ylabel('Income in 25th and 75th percentage ($)')

for bar in bar1 + bar2:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, f"{int(height/1000)}K", ha='center', va='bottom', fontsize=7)
    
plt.xticks(x + bar_width / 2, categories)

plt.legend(fontsize=7)
plt.show()
```

```python

data['Employing Rate'] = (data['Grad_employed'] / data['Grad_total']) * 100

grouped_data = data.groupby('Major_category')['Employing Rate'].mean()
grouped_data = grouped_data.sort_values(ascending=False)

plt.figure(figsize=(8, 5))
plt.bar(grouped_data.index, grouped_data)

plt.xlabel('Major Category')
plt.ylabel('Employing Rate (%)')
plt.title('Employing Rate of Grads by Major Category')

for i, value in enumerate(grouped_data):
    plt.text(i, value, f"{int(value)}%", ha='center', va='bottom')

plt.xticks(rotation='vertical', size=8)
plt.ylim(0, 100)
plt.gca().yaxis.set_major_formatter(ticker.PercentFormatter())

plt.show()
```

```python
grouped_data = data.groupby('Major').agg({'Grad_total': 'sum', 'Grad_median': 'median'})

plt.figure(figsize=(10, 8))

total_grads = grouped_data['Grad_total']
median_salary = grouped_data['Grad_median']

plt.scatter(total_grads, median_salary)

plt.xlabel('Total Number of Graduates')
plt.ylabel('Median Salary ($)')
plt.title('Relationship between Graduates and Median Salary')

plt.ticklabel_format(style='plain')
plt.show()
```
